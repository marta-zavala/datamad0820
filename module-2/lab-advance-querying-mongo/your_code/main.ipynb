# Advanced Querying Mongo

Importing libraries and setting up connection

```python
from pymongo import MongoClient
client = MongoClient("mongodb://localhost/companies") 
db = client.get_database()
```

### 1. All the companies whose name match 'Babelgum'. Retrieve only their `name` field.

```python
query1 = {"name":{"$eq":"Babelgum"}}
sol1 = list(db.companies.find(query1,{"_id":False,"name":True}))
print(sol1)
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
query2 = {"number_of_employees":{"$gt":5000}}
sol2 = list(db.companies.find(query2,{'number_of_employees':True, "name":True}).sort(("number_of_employees"),1).limit(20))
print(sol2)
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
query3 = {"$and":[{'founded_year':{"$gte": 2000 }},{'founded_year':{"$lte": 2005 }}]}
sol3 = list(db.companies.find(query3,{"_id":0,"name":1,"founded_year":1}).limit(5))
print(sol3)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
query4 = {"$and":[{"ipo.valuation_amount":{"$gte":100000000}}, {"founded_year":{"$gte":2000}}]}
list(db["companies"].find(query4,{"_id":0,"name":1,"ipo.valuation_amount":1}).sort("ipo.valuation_amount",-1).limit(5))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
query5 = {"$and":[{'founded_year':{"$lte": 2005 }},{'number_of_employees':{"$lte": 1000 }}]}
list(db.companies.find(query5,{'founded_year':True, "name":True}).sort(("number_of_employees"),-1))
```
